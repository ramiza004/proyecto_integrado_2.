## Creando un DataFrame de Notas Escolares Simulado en Python

**Entendiendo el Problema:**

Queremos generar un conjunto de datos ficticio que simule las notas de estudiantes de secundaria en Argentina. Este dataset será la base para crear un gráfico de barras que muestre el promedio de cada materia y la relación con las faltas.

**Herramientas que Usaremos:**

* **Pandas:** Para crear y manipular DataFrames.
* **NumPy:** Para generar números aleatorios.
* **Matplotlib:** Para crear el gráfico.

**Código Python:**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generando datos aleatorios
np.random.seed(42)  # Para reproducibilidad
materias = ['Matemática', 'Lengua', 'Ciencias Naturales', 'Historia', 'Geografía', 'Inglés', 'Educación Física']
nombres = ['Juan', 'María', 'Pedro', 'Ana', 'Lucas', 'Sofía', 'Camila', 'Diego', 'Martín', 'Laura']
apellidos = ['González', 'Pérez', 'Rodríguez', 'Fernández', 'López', 'Martínez', 'Díaz', 'Sánchez', 'Moreno', 'Castillo']

def generar_alumno():
    return {
        'Apellido': np.random.choice(apellidos),
        'Nombre': np.random.choice(nombres),
        'DNI': np.random.randint(10000000, 99999999),
        'Materia': np.random.choice(materias),
        'Nota1': np.random.randint(1, 11),
        'Nota2': np.random.randint(1, 11),
        'Nota3': np.random.randint(1, 11),
        'Faltas': np.random.randint(0, 20)
    }

# Creando el DataFrame
data = [generar_alumno() for _ in range(100)]
df = pd.DataFrame(data)

# Calculando promedios y agregando al DataFrame
df_promedios = df.groupby('Materia')[['Nota1', 'Nota2', 'Nota3', 'Faltas']].mean().reset_index()

# Creando el gráfico
plt.figure(figsize=(12, 6))
plt.bar(df_promedios['Materia'], df_promedios['Nota1'])
plt.xlabel('Materia')
plt.ylabel('Promedio')
plt.title('Promedio por Materia')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Agregando las faltas como un segundo eje
ax2 = plt.twinx()
ax2.plot(df_promedios['Materia'], df_promedios['Faltas'], color='red', marker='o')
ax2.set_ylabel('Faltas', color='red')

plt.legend(['Promedio', 'Faltas'])
plt.show()
```
